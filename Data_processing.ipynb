```python
import csv
import datetime
import pandas as pd
import numpy as np

# load data
df = pd.read_csv("mlbbatting1901-2021.csv")
df.Date = df.Date.apply(lambda x: datetime.datetime.strptime(x[:4] + x[5:7] + x[8:10], '%Y%m%d'))

print(df)
```

```python
# drop "Deadball Era" data & players with PA less than 100
c_deadball = datetime.datetime.strptime("1920-01-01", "%Y-%m-%d")
df_filtered = df.drop(df[df.Date < c_deadball].index, axis=0)
df_filtered = df_filtered.groupby('ID').filter(lambda x: x['PA'].sum() >= 100)

df_filtered.head(10)
```

```python
# Win or Lose
df_filtered.Rslt = df_filtered.Rslt.apply(lambda x: x[0] == "W")

df_filtered.head(10)
```

```python
# select columns
df_selected = df_filtered.loc[:, ["ID", "Player", "Date", "Tm", "Rslt", "PA", "AB", "R", "H", "2B", "3B", "HR", "RBI", "BB", "SO", "HBP", "SH", "SB", "BOP", "SF", "Pos Summary", "aLI", "WPA"]]
df_selected["SF"] = df_selected["SF"].replace(np.nan, 0)

df_selected.head(10)
```

```python
# WPA for High-Leverage PA

high_leverage_threshold = 1.5
df_high_leverage = df_selected[df_selected['aLI'] > high_leverage_threshold]

df_high_leverage = df_high_leverage.groupby(["ID"]).agg(WPA_total=("WPA", "sum"), High_Leverage_PA_total=("PA", "sum")).reset_index()
df_high_leverage["WPA_per_High_Leverage_PA"] = df_high_leverage["WPA_total"] / df_high_leverage["High_Leverage_PA_total"]

df_high_leverage.head(10)
```

```python
# Streakiness Score

df_streakiness_score = df_filtered.loc[:, ["ID", "Date", "H"]].sort_values(['ID', 'Date'])
df_streakiness_score['Hits_Rolling_Std'] = df_streakiness_score.groupby('ID')['H'].transform(lambda x: x.rolling(10, min_periods=1).std())
df_streakiness_score = df_streakiness_score.groupby('ID')['Hits_Rolling_Std'].mean().reset_index()

df_streakiness_score.head(10)
```

```python
# group data by ID and 
df_grouped = df_selected.groupby(["ID", "Player"]).agg(Win=("Rslt", "sum"), Game=("Rslt", "count"), PA=("PA", "sum"), AB=("AB", "sum"), Run=("R", "sum"), Hit=("H", "sum"), TwoBase=("2B", "sum"), ThreeBase=("3B", "sum"), HomeRun=("HR", "sum"), BB=("BB", "sum"), SF=("SF", "sum"), HBP=("HBP", "sum"), SB=("SB", "sum"), Date_min=("Date", "min"), Date_max=("Date", "max"))
df_grouped = df_grouped.rename(columns={"Hit": "H", "TwoBase": "2B", "ThreeBase": "3B", "HomeRun": "HR"})
df_grouped = df_grouped.reset_index()

df_grouped.head(10)
```

```python
# Add time info to player name
df_time_added = df_grouped
df_time_added["Player"] = df_grouped.apply(lambda x: x.Player + "(" + x.Date_min.strftime("%Y") + "~" + x.Date_max.strftime("%Y") + ")", axis=1)

df_time_added.head(10)
```

```python
# calculate various metrics
df_calculated = df_time_added

df_calculated["1B"] = df_calculated["H"] - df_calculated["2B"] - df_calculated["3B"] - df_calculated["HR"]
df_calculated["XBH"] = df_calculated["2B"] + df_calculated["3B"] + df_calculated["HR"]
df_calculated["AVG"] = df_calculated["H"] / df_calculated["AB"]
df_calculated["SLG"] = (df_calculated["1B"] + 2 * df_calculated["2B"] + 3 * df_calculated["3B"] + 4 * df_calculated["HR"]) / df_calculated["AB"]
df_calculated["OBP"] = (df_calculated["H"] + df_calculated["BB"] + df_calculated["HBP"]) / (df_calculated["AB"] + df_calculated["BB"] + df_calculated["HBP"] + df_calculated["SF"])
df_calculated["OPS"] = df_calculated["SLG"] + df_calculated["OBP"]
df_calculated["Hard_Hit_Efficiency"] = df_calculated["XBH"] / df_calculated["AB"]
df_calculated["Speed_Score"] = (df_calculated["SB"] + df_calculated["2B"] + df_calculated["3B"]) / df_calculated["PA"]
df_calculated["Runs_per_PA"] = df_calculated["Run"] / df_calculated["PA"]

df_calculated["1B_per_PA"] = df_calculated["1B"] / df_calculated["PA"]
df_calculated["2B_per_PA"] = df_calculated["2B"] / df_calculated["PA"]
df_calculated["3B_per_PA"] = df_calculated["3B"] / df_calculated["PA"]
df_calculated["HR_per_PA"] = df_calculated["HR"] / df_calculated["PA"]
df_calculated["BB_per_PA"] = df_calculated["BB"] / df_calculated["PA"]
df_calculated["SB_per_PA"] = df_calculated["SB"] / df_calculated["PA"]
df_calculated["HBP_per_PA"] = df_calculated["HBP"] / df_calculated["PA"]
df_calculated["SF_per_PA"] = df_calculated["SF"] / df_calculated["PA"]

df_calculated.head(10)
```

```python
# join dataframes
df_joined = df_calculated.merge(df_high_leverage, on="ID").merge(df_streakiness_score, on="ID")

df_joined.head(10)
```

```python
# write csv

df_joined.to_csv(path_or_buf="data_processed.csv", index=False)
```
